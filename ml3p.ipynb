```python
X=df.drop(columns=["price"],axis =1)
```

```python
x_encoded =pd.get_dummies(X ,columns=["model","transmission","fuelType"],drop_first=True)
x_encoded = x_encoded.astype(int)
```

```python
x_encoded
```

```python
from sklearn.preprocessing import LabelEncoder

columns = ["model", "transmission", "fuelType"]
xlable = X.copy()
label_encoders = {}

for col in columns:
    le = LabelEncoder()
    xlable[col] = le.fit_transform(xlable[col])
    label_encoders[col] = le 
```

```python
xlable
```

```python
x_encoded
```

```python
from sklearn.preprocessing import StandardScaler
cols = ["year","mileage","tax","mpg","engineSize"]
scaler = StandardScaler()
x_encoded[cols] = scaler.fit_transform(x_encoded[cols])
```

```python
x_encoded
```

```python
columns_to_scale = ['year', 'mileage', 'tax', 'mpg', 'engineSize']
scaler = StandardScaler()
xlable[columns_to_scale] = scaler.fit_transform(xlable[columns_to_scale])
```

```python
xlable
```

```python
from sklearn.linear_model import LinearRegression
from sklearn.model_selection import train_test_split
from sklearn.metrics import mean_squared_error, r2_score
y=df["price"]
# Train-test split
X_train, X_test, y_train, y_test = train_test_split(x_encoded, y, test_size=0.20, random_state=42)

# Initialize and train Linear Regression model
model = LinearRegression()
model.fit(X_train, y_train)

# Make predictions
y_pred = model.predict(X_test)
y_pred
```

```python
r2 = r2_score(y_test, y_pred)
r2
```

```python
n = X_test.shape[0]
p = X_test.shape[1]
adjusted_r2 =1-((1-r2) *(n-1)/ (n-p-1))
adjusted_r2
```
